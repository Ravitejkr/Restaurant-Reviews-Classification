```python
import pandas as pd
A = pd.read_csv("C:/Users/Admin/Desktop/Ravitej Kaur/Projects/Restaurant reviews classification/Restaurant_Reviews.tsv",sep='\t')
A.head()
```

```python
Q = []
from re import sub
for i in A.Review:
    Q.append(sub("[^a-zA-Z0-9]","",i.upper()))
```

```python
Q
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
words_vect = cv.fit_transform(Q).toarray()
```

```python
words = cv.get_feature_names()
```

```python
from sklearn.model_selection import train_test_split
X = words_vect
Y = A.Liked
xtrain,  xtest, ytrain, ytest = train_test_split(X,Y,test_size=0.2, random_state=21)
```

```python
X.shape
```

```python
xtrain.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
nn = Sequential()
nn.add(Dense(1000,input_dim=(995)))
nn.add(Dropout(0.8))
nn.add(Dense(1000))
nn.add(Dropout(0.8))
nn.add(Dense(1,activation="sigmoid"))
```

```python
nn.compile(loss="binary_crossentropy", metrics="accuracy")
nn.fit(xtrain, ytrain, epochs=10)
```

```python
Q = []
for i in nn.predict(xtest):
    if(i[0]<0.5):
        Q.append(0)
    else:
        Q.append(1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,Q)
```
